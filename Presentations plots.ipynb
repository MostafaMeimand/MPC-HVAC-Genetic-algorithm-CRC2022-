```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
import statistics
import re
```

```python
# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(15, 50, 0.01)
  
# Calculating mean and standard deviation  
plt.plot(x_axis, norm.pdf(x_axis, 31.42, 5.06),linewidth=3)
plt.plot(x_axis, norm.pdf(x_axis, 28.47, 4.50),linewidth=3)
plt.plot(x_axis, norm.pdf(x_axis, 30.37, 1.81),linewidth=3)
plt.axvline(x = 29.02,color = "red", linewidth = 2, linestyle='-.')
plt.legend(["One occupant","Two occupants","Four occupants","Conventional MPC"])
plt.xlabel("Energy consumption (kWh)")
plt.ylabel("Probability")
plt.yticks([])
# plt.savefig("EnergyConsumption.svg")
```

```python
## Results and discussion
temp_1 = pd.read_csv("/content/drive/Shareddrives/_Research Repository - Mostafa Meimand/Working Papers/Conferences/CRC 2022/eta10_subject_6.csv")
temp_1 = temp_1.drop(temp_1.index[0:192],axis = 0)
temp_1 = temp_1.reset_index()
temp_1 = temp_1.drop("index", axis = 1)

# Adding dates to the dataset
delimiters = " ", ":", "/"
regexPattern = '|'.join(map(re.escape, delimiters))
temp_1["Month"] = None
temp_1["Day"] = None
temp_1["Hour"] = None
temp_1["Minutes"] = None

for i in range(temp_1.shape[0]):
  temp_1["Month"][i] = int(re.split(regexPattern,temp_1["Date/Time"][i])[1])
  temp_1["Day"][i] = int(re.split(regexPattern,temp_1["Date/Time"][i])[2])
  temp_1["Hour"][i] = int(re.split(regexPattern,temp_1["Date/Time"][i])[4])
  temp_1["Minutes"][i] = int(re.split(regexPattern,temp_1["Date/Time"][i])[5])

temp_1 = temp_1[temp_1["Day"] == 1]

temp_1["CENTRAL SYSTEM_UNIT1:Air System Electricity Energy [J](TimeStep)"] *= 2.77778e-7

temp_1["time"] = temp_1.index
```

```python
fig,ax = plt.subplots()
ax.plot(temp_1['LIVING_UNIT1:Zone Mean Air Temperature [C](TimeStep)'].groupby(np.arange(96)//4).mean(), color="#6E7B8B", marker="o")
ax.plot(temp_1['LIVING_UNIT1:Zone Thermostat Cooling Setpoint Temperature [C](TimeStep)'].groupby(np.arange(96)//4).mean(), color="#87CEFA", marker="o")
ax.set_xlabel("hour",fontsize=14)
ax.set_ylabel("Temperature",fontsize=14)
ax.legend(["Indoor temperature","Setpoint"],loc='upper left')
ax.set_xticks(range(0,26,2))
ax.set_yticks(np.arange(22,28,1))
ax.axhline(y = 24.40, linestyle = "-.")
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
ax2.plot(temp_1["Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)"].groupby(np.arange(96)//4).mean(),color="#03A89E",marker="o")
ax2.tick_params(labelcolor="#03A89E")
ax2.set_ylabel("Outdoor temperature", fontsize = 14, color = "#03A89E")
# plt.savefig("Example for subject #6.svg")
```
